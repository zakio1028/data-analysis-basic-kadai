```python
import numpy as np
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 2, tight_layout=True)

#折れ線グラフのデータを設定
y0 = [1, 2, -5, 2]
#折れ線グラフを描画
axes[0,0].plot(y0)

#sin関数を設定する。
x = np.linspace(0, 10, 100)
y = 2 + 2 * np.sin(1 * x)

#sin関数を描画
axes[0,1].plot(x, y)

#ヒストングラフのデータを設定
x1 = np.random.normal(15, 5, 2000)
#ヒストングラフを描画
axes[1,0].hist(x1, bins=10)

# 散布図のデータを生成する
np.random.seed(3)
x2 = 4 + np.random.normal(0, 2, 60)
y2 = 4 + np.random.normal(0, 2, len(x2))

#散布図を描画
axes[1,1].scatter(x2, y2)
```
