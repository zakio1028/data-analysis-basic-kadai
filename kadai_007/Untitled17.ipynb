```python
import matplotlib.pyplot as plt
import pandas as pd
import japanize_matplotlib

# データ読み込みとマージ
df = pd.read_csv("sample_pandas_6.csv")
category_df = pd.read_csv('category.csv')
df = pd.merge(df, category_df[['商品番号','カテゴリー']], how='inner', on='商品番号')

# カテゴリーごとの商品数（棒グラフ）
cont_df = df['カテゴリー'].value_counts()
cont_df.plot(kind='bar', title='カテゴリーごとの商品数')
plt.ylabel('商品数')
plt.grid(True)
plt.show()

# 図を表示
plt.show()

df.groupby('カテゴリー')['単価'].describe()
```
